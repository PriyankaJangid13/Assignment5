```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from statsmodels.graphics.regressionplots import influence_plot
```

```python
data=pd.read_csv("50_Startups.csv")
data.head()
```

# EDA

```python
data.info()
```

```python
data.describe()
```

```python
data1=data.rename({'R&D Spend':'RDS','Administration':'ADM','Marketing Spend':'MS'},axis=1)
data1.head()
```

```python
df=data1.drop('State',axis=1).reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

# Correlation & visualiazation

```python
df.corr()  
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(df)
plt.show()
```

# Preparing The Model

```python
model=smf.ols('Profit~RDS+ADM+MS',data=df).fit()
```

```python
model.summary()
```

```python
model.params
```

```python
model.tvalues , np.round(model.pvalues,5)
```

```python
model.rsquared , model.rsquared_adj
```

```python
# Build SLR and MLR models for insignificant variables 'ADMS' and 'MKTS'
# Also find their tvalues and pvalues
```

```python
slr_a=smf.ols("Profit~ADM",data=df).fit()
slr_a.tvalues , slr_a.pvalues  # ADM has in-significant pvalue
```

```python
slr_m=smf.ols("Profit~MS",data=df).fit()
slr_m.tvalues , slr_m.pvalues  # MS has significant pvalue
```

```python
mlr=smf.ols("Profit~ADM+MS",data=df).fit()
mlr.tvalues , mlr.pvalues  # varaibles have significant pvalues
```

# Model Validation

```python
rsq_rds=smf.ols('RDS~ADM+MS',data=df).fit().rsquared
vif_rds=1/(1-rsq_rds)
rsq_adm=smf.ols('ADM~RDS+MS',data=df).fit().rsquared
vif_adm=1/(1-rsq_adm)
rsq_ms=smf.ols('MS~RDS+ADM',data=df).fit().rsquared
vif_ms=1/(1-rsq_ms)
d1={'variables':['RDS','ADM','MS'],'vif':[vif_rds,vif_adm,vif_ms]}
df1=pd.DataFrame(d1)
df1
```

# QQ plot

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title("QQ-plot")
plt.show()
```

# Residual Plot

```python
def get_standardized_values( vals ):
      return ( vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),get_standardized_values(model.resid))
plt.title('Residual plot')
plt.xlabel('Fitted values')
plt.ylabel('Residual values')
plt.show()
```

# Residual vs Regressor plot

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'ADM',fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'RDS',fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'MS',fig)
plt.show()
```
