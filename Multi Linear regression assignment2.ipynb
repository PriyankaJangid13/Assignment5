```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+qt+Weight',data=car5).fit()
model
```

```python
model.params
```

```python
model.tvalues,np.round(model.pvalues,4)
```

```python
model.rsquared,model.rsquared_adj
```

```python
slr_cc=smf.ols('Price~cc',data=car5).fit()
slr_cc.tvalues,slr_cc.pvalues
```

```python
slr_dd=smf.ols('Price~Doors',data=car5).fit()
slr_dd.tvalues,slr_dd.pvalues
```

```python
mlr_cd=smf.ols('Price~cc+Doors',data=car5).fit()
mlr_cd.tvalues,mlr_cd.pvalues
```

```python
rsq_age=smf.ols('Age~KM+HP+cc+Doors+Gears+qt+Weight',data=car5).fit().rsquared
vif_age=1/(1-rsq_age)
rsq_km=smf.ols('KM~Age+HP+cc+Doors+Gears+qt+Weight',data=car5).fit().rsquared
vif_km=1/(1-rsq_km)
rsq_hp=smf.ols('HP~Age+KM+cc+Doors+Gears+qt+Weight',data=car5).fit().rsquared
vif_hp=1/(1-rsq_hp)
rsq_cc=smf.ols('cc~Age+KM+HP+Doors+Gears+qt+Weight',data=car5).fit().rsquared
vif_cc=1/(1-rsq_cc)
rsq_doors=smf.ols('Doors~Age+KM+HP+cc+Gears+qt+Weight',data=car5).fit().rsquared
vif_doors=1/(1-rsq_doors)
rsq_gears=smf.ols('Gears~Age+KM+HP+cc+Doors+qt+Weight',data=car5).fit().rsquared
vif_gears=1/(1-rsq_gears)
rsq_qt=smf.ols('qt~Age+KM+HP+cc+Doors+Gears+Weight',data=car5).fit().rsquared
vif_qt=1/(1-rsq_qt)
rsq_weight=smf.ols('Weight~Age+KM+HP+cc+Doors+Gears+qt',data=car5).fit().rsquared
vif_weight=1/(1-rsq_weight)
d1={'veriable':['Age','KM','HP','cc','Doors','Gears','qt','Weight'],'vif':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_qt,vif_weight]}
vif_df=pd.DataFrame(d1)
vif_df
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
sm.qqplot(model.resid,line='q')
plt.title('Q-Q plot')
```

```python
def get_standaridized_Values(vals):return(vals-vals.mean())/vals.std()
plt.scatter(get_standaridized_Values(model.fittedvalues),get_standaridized_Values(model.resid))
plt.title('residual plot')
plt.xlabel('standerdized filted values')
plt.ylabel('standardized residuel values')
plt.show()
```
